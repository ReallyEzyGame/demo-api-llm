# Run Test

```python
import subprocess
import sys
import time

# Mở file log để ghi
log_file = open("server.log", "w")

# Khởi chạy server tách biệt hoàn toàn khỏi vòng đời của Cell
process = subprocess.Popen(
    [sys.executable, "./source/api.py"],
    stdout=log_file,
    stderr=subprocess.STDOUT,
    start_new_session=True  # Cực kỳ quan trọng: Ngăn Colab tự động kill process khi cell kết thúc
)

print(f"🚀 Tiến trình Server đang chạy ngầm (PID: {process.pid})")
print("⏳ Đang tải Model vào GPU, quá trình này mất khoảng 1-2 phút...\n")
print("-" * 50)

# Tự động đọc file log để thông báo khi xong
with open("server.log", "r") as f:
    while True:
        line = f.readline()
        if not line:
            time.sleep(0.5)
            continue

        # In log ra màn hình để bạn theo dõi tiến độ
        print(line, end="")

        # Nhận diện dấu hiệu Server đã sẵn sàng
        if "Application startup complete" in line:
            print("-" * 50)
            print("✅ SERVER ĐÃ SẴN SÀNG!")
            print("Bạn có thể dừng chạy cell này (bấm nút Stop hình vuông) và chạy file localTest.py ở cell tiếp theo.")
            break

        # Nhận diện dấu hiệu bị tắt đột ngột
        if "Shutting down" in line:
            print("-" * 50)
            print("❌ LỖI: Server đã bị tắt!")
            break
```
